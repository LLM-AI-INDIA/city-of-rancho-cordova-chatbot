```python
import os
import numpy as np
from urllib.parse import urljoin, urlparse
from bs4 import BeautifulSoup
from collections import deque

import faiss
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

print("All libraries loaded.")
```

```python
BASE_URL = "https://www.cityofranchocordova.org/"
```

```python
import asyncio
from playwright.async_api import async_playwright
from urllib.parse import urljoin, urlparse
from collections import deque
from bs4 import BeautifulSoup

def is_same_domain(base, new):
    return urlparse(base).netloc == urlparse(new).netloc

def extract_visible_text(html):
    soup = BeautifulSoup(html, "html.parser")
    for s in soup(["script","style","noscript","iframe"]):
        s.extract()
    return soup.get_text(" ", strip=True)

async def crawl_with_browser(start_url, max_pages=40):
    pages = {}
    visited = set()
    queue = deque([start_url])

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        while queue and len(visited) < max_pages:
            url = queue.popleft()
            if url in visited:
                continue

            print("Visiting:", url)

            try:
                await page.goto(url, timeout=60000, wait_until="load")
                html = await page.content()
                visited.add(url)

                text = extract_visible_text(html)
                if len(text) > 100:
                    pages[url] = text

                soup = BeautifulSoup(html, "html.parser")
                for a in soup.find_all("a", href=True):
                    next_url = urljoin(url, a["href"])
                    if is_same_domain(start_url, next_url):
                        queue.append(next_url)

            except Exception as e:
                print("Error on:", url, e)

        await browser.close()

    return pages
```

```python
BASE_URL = "https://www.cityofranchocordova.org/"

pages = await crawl_with_browser(BASE_URL, max_pages=40)

print("Total pages crawled:", len(pages))
list(pages.keys())[:10]
```

```python
documents = []
sources = []

for url, text in pages.items():
    words = text.split()
    if len(words) < 50:
        continue

    for i in range(0, len(words), 200):
        chunk = " ".join(words[i:i+200])
        if len(chunk.strip()) > 0:
            documents.append(chunk)
            sources.append(url)

print("Chunks created:", len(documents))
```

```python
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = embed_model.encode(documents, convert_to_numpy=True)
dim = embeddings.shape[1]

index = faiss.IndexFlatL2(dim)
index.add(embeddings)

print("Embeddings:", embeddings.shape)
print("FAISS index ready.")
```

```python
model_name = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
llm = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
def search(query, k=5):
    q_emb = embed_model.encode([query], convert_to_numpy=True)
    dist, idxs = index.search(q_emb, k)
    return [(documents[i], sources[i]) for i in idxs[0]]

def answer_question(query):
    retrieved = search(query)

    context = "\n\n".join(
        [f"Source: {src}\n{txt}" for txt, src in retrieved]
    )

    prompt = f"""
    Use the context below to answer the user question.

    Context:
    {context}

    Question: {query}

    Give a short, accurate answer.
    """

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)
    output = llm.generate(**inputs, max_new_tokens=200)
    answer = tokenizer.decode(output[0], skip_special_tokens=True)
    return answer
```

```python
answer_question("What services does the city provide?")
```

```python
import pickle

pickle.dump(documents, open("documents.pkl","wb"))
pickle.dump(sources, open("sources.pkl","wb"))
faiss.write_index(index, "faiss.index")

print("Saved.")
```

```python
%%writefile app.py
import streamlit as st
import pickle
import faiss
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import numpy as np

documents = pickle.load(open("documents.pkl","rb"))
sources = pickle.load(open("sources.pkl","rb"))
index = faiss.read_index("faiss.index")

embed_model = SentenceTransformer("all-MiniLM-L6-v2")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
llm = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")

def search(query, k=5):
    q = embed_model.encode([query], convert_to_numpy=True)
    dist, idxs = index.search(q, k)
    return [(documents[i], sources[i]) for i in idxs[0]]

def answer_question(query):
    retrieved = search(query)

    context = "\n\n".join(
        [f"Source: {s}\n{t}" for t,s in retrieved]
    )

    prompt = f"Context:\n{context}\n\nQuestion: {query}\n\nAnswer:"

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)
    out = llm.generate(**inputs, max_new_tokens=200)
    ans = tokenizer.decode(out[0], skip_special_tokens=True)
    return ans

st.title("Rancho Cordova — Website QA Chatbot (Open Source RAG)")

q = st.text_input("Ask a question:")
if q:
    st.write("### Answer:")
    st.write(answer_question(q))
```
